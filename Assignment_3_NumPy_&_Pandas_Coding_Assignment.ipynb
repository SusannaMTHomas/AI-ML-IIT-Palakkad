##Q3. Matrix Operations for Business Data

A company has sales data for 3 products across 4 regions.
<br>sales_matrix = np.array([
<br>[120,140,130,150],
<br>[80,90,100,95],
<br>[200,210,190,205]
])

###1. Find total sales per product.

```python
sales_matrix = np.array([
[120,140,130,150],
[80,90,100,95],
[200,210,190,205] ])
sales_product=np.sum(sales_matrix,axis=1)
print("Total sales per product: ",sales_product)
```

###2. Find total sales per region.

```python
sales_region=np.sum(sales_matrix,axis=0)
print("Total sales per region: ",sales_region)
```

###3. Identify the best performing product.

```python
best_product = np.argmax(sales_product)+1  #+1 as indexing starts at zero
print("Best performing product: ",best_product)
```

###4. Identify the region with highest sales.

```python
best_region = np.argmax(sales_region)+1
print("Region with highest sales: ",best_region)
```

##Q5. Missing Data Handling
<br>Dataset:
<br>data = {
<br>'Name':['John','Anna','Peter','Linda','David'],
<br>'Age':[28, np.nan, 35, 32, np.nan],
<br>'Salary':[50000,54000,np.nan,58000,60000]
}


###1. Detect missing values.


```python
data = {
'Name':['John','Anna','Peter','Linda','David'],
'Age':[28, np.nan, 35, 32, np.nan],
'Salary':[50000,54000,np.nan,58000,60000] }
#creating  a dataframe
emp_df = pd.DataFrame(data)
emp_df
```

```python
emp_df.isnull()
```

```python
#to find count of missing values
emp_df.isnull().sum()
```

###2. Replace missing Age with mean age.

```python
emp_df['Age'].fillna(emp_df['Age'].mean(),inplace=True)
emp_df
```

###3. Replace missing Salary with median salary.

```python
emp_df['Salary'].fillna(emp_df['Salary'].median(),inplace=True)
emp_df
```

###4. Remove rows with any remaining missing values.

```python
emp_df.dropna(axis=0,inplace=True)
emp_df
```

##Q7. Customer Purchase Analysis
<br>Dataset:
<br>data = {
<br>'CustomerID':[101,102,103,104,105,106,107],
<br>'Age':[25,45,35,50,23,40,60],
<br>'PurchaseAmount':[250,600,400,800,150,700,900],
<br>'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}

```python
data = {
'CustomerID':[101,102,103,104,105,106,107],
'Age':[25,45,35,50,23,40,60],
'PurchaseAmount':[250,600,400,800,150,700,900],
'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai'] }
#creating  a dataframe
customer_df = pd.DataFrame(data)
customer_df
```

###1. Find the average purchase amount by city.

```python
customer_df.groupby('City')['PurchaseAmount'].mean()
```

###2. Identify customers with purchase above overall average.

```python
Avg_purchase= customer_df['PurchaseAmount'].mean()
above_avg_customer_df=customer_df[customer_df['PurchaseAmount']>Avg_purchase]
above_avg_customer_df
```

```python
above_avg_customer_df['CustomerID']
```

###3. Find the city generating the highest revenue.

```python
top_city_sales=customer_df.groupby('City')['PurchaseAmount'].sum()
top_city=top_city_sales.idxmax()
print("City generating highest revenue: ",top_city)
```

###4. Create a column Age_Group:
<br><30 → Young
<br>30–50 → Adult
<br>50 → Senior

```python
#defining a function for creating age_group
def age_group(age):
  if age<30:
    return 'Young'
  elif (age>=30 and age<50):
    return 'Adult'
  else:
    return 'Senior'

customer_df['Age_group']=customer_df['Age'].apply(age_group)
customer_df
```

###5. Find average purchase per age group.

```python
customer_df.groupby('Age_group')['PurchaseAmount'].mean()
```

##Q8. Time Series Sales Analysis
<br>Generate data:
<br>dates = pd.date_range('2024-01-01', periods=12, freq='M')
<br>sales = [200,220,250,270,300,320,350,360,380,400,420,450]
<br>df = pd.DataFrame({'Date':dates,'Sales':sales})


```python
dates = pd.date_range('2024-01-01', periods=12, freq='M')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date':dates,'Sales':sales})
df
```

###1. Convert Date to datetime index.


```python
df.set_index('Date',inplace=True)
df
```

###2. Compute monthly percentage growth.

```python
#group data by month and gets total sum
monthly_sales = df['Sales'].resample('M').sum()
monthly_prcntg_growth = monthly_sales.pct_change()*100
monthly_prcntg_growth
```

###3. Calculate 3-month rolling average sales.

```python
rolling_avg=df['Sales'].rolling(window=3).mean()
rolling_avg
```

###4. Identify the month with highest growth rate.

```python
Highest_growth_month = monthly_prcntg_growth.idxmax()
print('Month with highest growth rate: ',Highest_growth_month.strftime("%B"),'-',Highest_growth_month.year)
```
