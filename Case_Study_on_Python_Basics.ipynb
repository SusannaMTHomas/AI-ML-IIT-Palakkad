##c. Implement conditional statements to check for sufficient balance before withdrawal

```python
#defining a function for withdrawing money
def withdraw_money():
  global acc_balance
  #defining min_balance
  min_balance=1000
  amount=float(input("Enter amount to withdraw: "))
  if(acc_balance-amount>=min_balance):
    acc_balance=acc_balance-amount
    print("Withdrawal successful.\nRemaining balance: Rs.",acc_balance)
  else:
    print("Insuffucient funds. You must maintain a minimum balance of Rs.",min_balance)
```

##a. Analyze weather data (temperature, humidity, etc.) for a given region


```python
#weather data
region ="Chennai"
temp_list = [32.2,32.4,33,33.5,32.8,31]
humidity_list=[65,64,63,61,62,66]
wind_speed_list=[10,12,15,11,14,16]
```

##b. Use variables and data types to store weather data

```python
#checking datatypes of variables
print("Data type of temp_list - ",type(temp_list))
print("Data type of humidity_list - ",type(humidity_list))
print("Data type of wind_speed_list - ",type(wind_speed_list))
```

##c. Implement conditional statements to determine weather conditions(sunny, rainy, etc.)

```python
#defining a function to describe weather condition
def weather_condition(avg_temp,av_humidity,avg_wind_speed):
  if(avg_temp)>32.5 and av_humidity<60:
    print("Weather condition is sunny.")
  elif av_humidity>72:
     print("Weather condition is rainy.")
  elif avg_wind_speed>12.5:
     print("Weather condition is windy.")
  else:
     print("Weather condition is cloudy.")
```

```python
#example scenario - case 1
weather_condition(33,59,11)
```

```python
#example scenario - case 2
weather_condition(32,75,11)
```

```python
#example scenario - case 3
weather_condition(32,65,13)
```

```python
#example scenario - case 4
weather_condition(32,65,12)
```
